```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

# Assuming the data from the picture has been saved as 'boston.csv'
# Load the data
file_path = '/content/boston.csv'  # Update this to the correct path
df = pd.read_csv(file_path)

# Display the DataFrame to verify the data
print(df.head())

# Separate features and target variable from the DataFrame
X = df.drop(columns=['PRICE'])
y = df['PRICE']

# Forward selection function
def forward_selection(X, y):
    initial_features = X.columns.tolist()
    best_features = []
    while len(initial_features) > 0:
        remaining_features = list(set(initial_features) - set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(y, sm.add_constant(X[best_features + [new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if min_p_value < 0.05:  # You can adjust the threshold as needed
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features

# Perform forward selection
selected_features = forward_selection(X, y)

# Print the selected features
print("Selected features:", selected_features)

# Fit the final model with selected features
final_model = sm.OLS(y, sm.add_constant(X[selected_features])).fit()
print(final_model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

file_path = '/content/sample.csv'  # Update this to the correct path
df = pd.read_csv(file_path)



# Define the dependent variable ('size' in this case)
Y = df['size']

# Define the independent variables (length, width, decimal, weight, height)
X = df[['length', 'width', 'decimal', 'weight', 'height']]

# Add a constant to the independent variables matrix
X = sm.add_constant(X)

# Perform bi-directional elimination using stepwise regression
def forward_backward_selection(X, Y, threshold_in=0.01, threshold_out=0.05, verbose=True):
    included = []
    history = []

    while True:
        changed = False

        # Forward step
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded, dtype=float)
        for new_column in excluded:
            model = sm.OLS(Y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            history.append((included.copy(), 'add', best_feature, best_pval))
            if verbose:
                print(f'Add {best_feature} with p-value {best_pval:.6f}')

        # Backward step
        model = sm.OLS(Y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max()
        if worst_pval > threshold_out:
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            changed = True
            history.append((included.copy(), 'drop', worst_feature, worst_pval))
            if verbose:
                print(f'Drop {worst_feature} with p-value {worst_pval:.6f}')

        if not changed:
            break

    # Plot the inclusion and exclusion process
    fig, ax = plt.subplots()
    actions = ['add' if action == 'add' else 'drop' for _, action, _, _ in history]
    p_values = [pval for _, _, _, pval in history]
    features = [feature for _, _, feature, _ in history]

    ax.plot(range(len(actions)), p_values, marker='o')
    for i, (action, feature) in enumerate(zip(actions, features)):
        ax.annotate(f"{action} {feature}", (i, p_values[i]))

    plt.xlabel('Step')
    plt.ylabel('p-value')
    plt.title('Bi-Directional Elimination Process')
    plt.show()

    return included

selected_features = forward_backward_selection(X, Y)
print('Selected features:', selected_features)

# Plot the p-values of the final selected features
model = sm.OLS(Y, sm.add_constant(pd.DataFrame(X[selected_features]))).fit()
pvalues = model.pvalues.iloc[1:]

plt.figure()
pvalues.plot(kind='bar', color='skyblue')
plt.xlabel('Features')
plt.ylabel('p-value')
plt.title('P-values of Final Selected Features')
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
file_path = '/content/sample.csv'
df = pd.read_csv(file_path)

# Display the DataFrame
print("DataFrame loaded from sample.csv:")
print(df)

# Perform some basic comparisons
# Calculate the average values for length, width, and size
average_length = df['length'].mean()
average_width = df['width'].mean()
average_size = df['size'].mean()

print(f"Average Length: {average_length}")
print(f"Average Width: {average_width}")
print(f"Average Size: {average_size}")

# Calculate standard deviation for length, width, and size
std_length = df['length'].std()
std_width = df['width'].std()
std_size = df['size'].std()

print(f"Standard Deviation Length: {std_length}")
print(f"Standard Deviation Width: {std_width}")
print(f"Standard Deviation Size: {std_size}")

# Filter entries where length is greater than 5
length_greater_than_5 = df[df['length'] > 5]
print("Entries with length greater than 5:")
print(length_greater_than_5)

# Identify and display the lowest value for length, width, and size
min_length = df['length'].min()
min_width = df['width'].min()
min_size = df['size'].min()

print(f"Lowest Length Value: {min_length}")
print(f"Lowest Width Value: {min_width}")
print(f"Lowest Size Value: {min_size}")

# Plot trends for length, width, and size
plt.figure(figsize=(10, 6))

plt.plot(df['length'], label='Length')
plt.plot(df['width'], label='Width')
plt.plot(df['size'], label='Size')

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Trends: Length, Width, Size')
plt.legend()
plt.grid(True)
plt.show()


```
